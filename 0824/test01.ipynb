```python
import csv
f = open('age.csv')
data = csv.reader(f)
result = []
for row in data:
    if '신도림' in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))
print(result)
```

```python
import csv
import matplotlib.pyplot as pit

f = open('age.csv')
data = csv.reader(f)
name = input('인구 구조가 알고싶은 지역의 이름(읍 면 동 단위) : ')
result = []
for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))

pit.style.use('ggplot')
pit.rc('font',family='Malgun Gothic')
pit.title(name + ' 지역의 인구 구조')
pit.plot(result)
pit.show()
```

```python
import matplotlib.pyplot as pit
import csv

f = open('age.csv')
data = csv.reader(f)
result = []
for row in data:
    if '신도림' in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))

pit.figure(figsize=(15,10))
#pit.bar(range(101),result)
pit.barh(range(101),result)
pit.show()
```

```python
import matplotlib.pyplot as pit
import csv

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
for row in data:
    if '신도림' in row[0]:
        for i in range(0,101):
            m.append(-int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse()
print(m)
print(f)
```

```python
import matplotlib.pyplot as pit

pit.barh(range(101),m)
pit.barh(range(101),f)
pit.show()
```

```python
# 성별로 대칭
import matplotlib.pyplot as pit

pit.rc('font',family='Malgun Gothic')
pit.title('신도림 지역의 성별 인구 분표')

pit.barh(range(101),m, label='남성')
pit.barh(range(101),f, label='여성')
pit.legend()
pit.show()
```

```python
import matplotlib.pyplot as pit
import csv

f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
name = input('찾고싶은 지역의 이름 : ')
for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(-int(i)) # 대칭으로 표시할 때
        for i in row[106:]:
            f.append(int(i))
pit.style.use('ggplot')
pit.figure(figsize=(10,5),dpi=300)
pit.rc('font',family='Malgun Gothic')
pit.rcParams['axes.unicode_minus'] = False
pit.title(name + ' 지역의 성별 인구 분표')

pit.barh(range(101),m, label='남성')
pit.barh(range(101),f, label='여성')
pit.legend()
pit.show()
```
