```python
# 산점도
import matplotlib.pyplot as pit
```

```python
pit.scatter([1,2,3,4],[10,30,20,40],s=[100,200,250,300])
pit.show()
```

```python
pit.scatter([1,2,3,4],[10,30,20,40],s=[100,200,250,300],c=range(4),cmap='jet')
pit.colorbar()
pit.show()
```

```python
import matplotlib.pyplot as pit
import random

x = []
y = []
size = []

for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))

pit.scatter(x,y, s=size,c=size,cmap='jet', alpha=0.8)
pit.colorbar()
pit.show()
```
