```python
# 1. 주어진 데이터 테이블 완성해서 출력
import pandas as pd
import numpy as np
df = pd.DataFrame({'id': ['1','2','3','4','5'],'kuk':[90,70,40,80,50],
                        'eng':[100,80,60,70,30],'math':[80,90,50,90,70]})
df
```

```python
# 2. 수학 점수가 90점 이상인 학생 출력

print(df['math']>=90)
```

```python
# 3. 총점 필드를 추가하여 출력
df['총점'] = (df['kuk']+df['eng']+df['math'])
df
```

```python
# 4.평균 필드를 추가하고 내림차순으로 출력
df['평균'] = (df['kuk']+df['eng']+df['math'])/3
df.sort_values('평균',ascending=False)
```
